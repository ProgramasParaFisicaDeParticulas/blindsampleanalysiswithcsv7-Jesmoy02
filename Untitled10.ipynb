```python
# Definir función para calcular masa invariante
def invariant_mass(pt1, eta1, phi1, pt2, eta2, phi2):
    """Función para calcular la masa invariante."""
    px1 = pt1 * np.cos(phi1)
    py1 = pt1 * np.sin(phi1)
    pz1 = pt1 * np.sinh(eta1)
    E1 = np.sqrt(px1**2 + py1**2 + pz1**2)

    px2 = pt2 * np.cos(phi2)
    py2 = pt2 * np.sin(phi2)
    pz2 = pt2 * np.sinh(eta2)
    E2 = np.sqrt(px2**2 + py2**2 + pz2**2)

    invariant_mass_squared = (E1 + E2)**2 - (px1 + px2)**2 - (py1 + py2)**2 - (pz1 + pz2)**2
    invariant_mass = np.sqrt(np.maximum(0, invariant_mass_squared))  # Avoid negative values due to precision issues

    return invariant_mass

# Calcular la masa invariante con los datos filtrados
filtered_charge.loc[:, 'invariant_mass'] = invariant_mass(filtered_charge['pt1'], filtered_charge['eta1'], filtered_charge['phi1'],
                                                          filtered_charge['pt2'], filtered_charge['eta2'], filtered_charge['phi2'])

# Calcular estadísticas para la masa invariante filtrada
statistics_mass = {
    'Count': filtered_charge['invariant_mass'].count(),
    'Media': filtered_charge['invariant_mass'].mean(),
    'Desviación Estándar': filtered_charge['invariant_mass'].std(),
    'Moda': filtered_charge['invariant_mass'].mode()[0],
    'Mínimo': filtered_charge['invariant_mass'].min(),
    'Máximo': filtered_charge['invariant_mass'].max()
}

# Mostrar estadísticas de la masa invariante
print("Estadísticas de la masa invariante:")
print(f"Count: {statistics_mass['Count']}")
print(f"Media: {statistics_mass['Media']}")
print(f"Desviación estándar: {statistics_mass['Desviación Estándar']}")
print(f"Mínimo: {statistics_mass['Mínimo']}")
print(f"Máximo: {statistics_mass['Máximo']}")

# Crear histograma de la masa invariante
plt.figure(figsize=(10, 6))
plt.hist(filtered_charge['invariant_mass'], bins=50, alpha=0.75, color='blue')
plt.title('Histograma de la masa invariante (datos filtrados)')
plt.xlabel('Masa invariante (GeV)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```
